# MI analysis of the qm7 extendended db with radical cations

The Mutual Information (**MI**) is estimated using the Gaussian Copula Mutual Information (**GCMI**) apporach between compressed versions of the decriptors and between compressed versions of the desciprots and the property. 

Despite the GCMI estimator is stable in high dimensions, the descriptors are extremely large compared to the dataset size, therefore some dimensionality reduction is necessary. I apply **PCA** and retain 99% of the variance to obtain a near-lossless compressed representation.

I **rank-Gaussianize** the descriptors before PCA so that dimensionality reduction reflects dependence structure rather than marginal scaling (just like the GCMI approach), then retain 99% variance.

To fascilitate the comparison, all the reported MI are nomalized in a range between 0 and 1. The noralizaiton is done by dividing teh MI by the minimum entropy of the components.


## Descriptors MI

### GLobal descriptors 

MI between global descriptors to analyze how many information they share in this DB.

```python

descriptor = ["a-SPAHM", "b-SPAHM", "SLATM"]



MI = np.empty([3, 3])

fig, (ax0, ax1, ax2) = plt.subplots(1,3, figsize=(12, 4))

fig.suptitle("Normalized MI between global descriptors", fontsize="x-large")

norm = mpl.colors.Normalize(vmin=0.0, vmax=1.0)
my_cmap="bwr"

### Neutral ##########################################
MI[0,1] = 0.598 # a-SPAHM vs b-SPAHM
MI[0,2] = 0.651 # a-SPAHM vs SLATM
MI[1,2] = 0.187 #b-SPAHM vs SLATM

for i in range(3):
    MI[i,i] = 1.0
MI[1,0] = MI[0,1]
MI[2,0] = MI[0,2]
MI[2,1] = MI[1,2]

im = ax0.imshow(MI,norm=norm, cmap=my_cmap)
# Show all ticks and label them with the respective list entries
ax0.set_xticks(range(len(descriptor)), labels=descriptor,
              rotation=45, ha="right", rotation_mode="anchor")
ax0.set_yticks(range(len(descriptor)), labels=descriptor)

# Loop over data dimensions and create text annotations.
for i in range(len(descriptor)):
    for j in range(len(descriptor)):
        text = ax0.text(j, i, MI[i, j],
                       ha="center", va="center", color="black")

ax0.set_title("Neutral")


### Optimized radical cation ##########################################
MI[0,1] = 0.537 # a-SPAHM vs b-SPAHM
MI[0,2] = 0.518 # a-SPAHM vs SLATM
MI[1,2] = 0.206 #b-SPAHM vs SLATM

for i in range(3):
    MI[i,i] = 1.0
MI[1,0] = MI[0,1]
MI[2,0] = MI[0,2]
MI[2,1] = MI[1,2]
im = ax1.imshow(MI,norm=norm, cmap=my_cmap)
# Show all ticks and label them with the respective list entries
ax1.set_xticks(range(len(descriptor)), labels=descriptor,
              rotation=45, ha="right", rotation_mode="anchor")
ax1.set_yticks([])

# Loop over data dimensions and create text annotations.
for i in range(len(descriptor)):
    for j in range(len(descriptor)):
        text = ax1.text(j, i, MI[i, j],
                       ha="center", va="center", color="black")
ax1.set_title("Optimized radical cation")

### Vertical radical cation ##########################################
MI[0,1] = 0.583 # a-SPAHM vs b-SPAHM
MI[0,2] = 0.565 # a-SPAHM vs SLATM
MI[1,2] = 0.230 #b-SPAHM vs SLATM

for i in range(3):
    MI[i,i] = 1.0
MI[1,0] = MI[0,1]
MI[2,0] = MI[0,2]
MI[2,1] = MI[1,2]
im = ax2.imshow(MI,norm=norm, cmap=my_cmap)
# Show all ticks and label them with the respective list entries
ax2.set_xticks(range(len(descriptor)), labels=descriptor,
              rotation=45, ha="right", rotation_mode="anchor")
ax2.set_yticks([])

# Loop over data dimensions and create text annotations.
for i in range(len(descriptor)):
    for j in range(len(descriptor)):
        text = ax2.text(j, i, MI[i, j],
                       ha="center", va="center", color="black")
ax2.set_title("Vertical radical cation")

cbar = fig.colorbar(im, ax=[ax0, ax1, ax2],
                    fraction=0.046, pad=0.04)
cbar.set_label("Mutual information")


#fig.tight_layout()
plt.show()
```

a-SPAHM share information with b-SPAHM and SLATM, but b-SPAHM and SLATM do not share much information for this db.

## Descriptors and SPE

```python
blocks = ["Neutral", "Optimized radical cation", "Vertical radical cation"]

mi_aSPAHM_global = [ 0.526, 0.724, 0.664]
mi_bSPAHM_global = [ 0.912, 0.996, 0.985]
mi_aSPAHM_local = [ 0.590, 0.664, 0.688]
mi_bSPAHM_local = [ 0.763, 0.851, 0.881]
mi_slatm = [ 0.889, 0.886, 0.890]

x = np.arange(len(blocks))
w = 0.15
s0 = 0.01

fig, ax = plt.subplots(figsize=(14, 4))

bar_container = ax.bar(x - (w*2) -s0, mi_aSPAHM_global, w, label="a-SPAHM (global)")
ax.bar_label(bar_container, padding=-15)
bar_container = ax.bar(x - w - s0, mi_aSPAHM_local, w, label="a-SPAHM (local)")
ax.bar_label(bar_container, padding=-15)
bar_container = ax.bar(x , mi_bSPAHM_global, w, label="b-SPAHM (global)")
ax.bar_label(bar_container, padding=-15)
bar_container = ax.bar(x + w, mi_bSPAHM_local, w, label="b-SPAHM (local)")
ax.bar_label(bar_container, padding=-15)
bar_container = ax.bar(x + (2*w) + s0, mi_slatm, w, label="SLATM")
ax.bar_label(bar_container, padding=-15)

ax.set_xticks(x, blocks)
ax.set_ylim(0,1)
ax.set_ylabel("Mutual Information")

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=False, shadow=False, ncol=5)
ax.set_title("Normalized MI between representations and SPE")
plt.show()
```

Despite having low MI among them, b-spahm and slatm show higher MI than a-SPAHM for this property. In all cases global b-SPAHM has higher MI, followed by SLATM, this differences is bigger for the optimized radical cations and lower for the neutral. Technically and information wise speaking, for this subsets, b-SPAHM has a bigger potential; nevertheless, the MI is not related with the fascility to retrieve and use the information contained, that is  directly related with the model and how it uses the available information.
